```python
import os
import matplotlib.pyplot as plt

import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

```python
dataset_path = 'dataset'
train_data_path = dataset_path + '/train'
test_data_path = dataset_path + '/test'
train_data_path, test_data_path
```

```python
classes = os.listdir(train_data_path)


print("===No of Images per class in training data===")
for cls in classes:
    imgs = os.listdir(train_data_path+'/'+cls)
    imgs_count = len(imgs)
    print(f"{cls} - {imgs_count} images")
```

```python
classes = os.listdir(test_data_path)

print("===No of Images per class in test data===")
for cls in classes:
    imgs_count = len(os.listdir(test_data_path+'/'+cls))
    print(f"{cls} - {imgs_count} images")
```

```python
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

train_dataset = datasets.ImageFolder(root=train_data_path, transform=train_transform)
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
train_dataset, train_loader
```

```python
i=1
for imgs, labels in train_loader:
    plt.subplot(3,3,i)
    i+=1
    plt.imshow(imgs[i-1].permute(1,2,0))
    if i == 10:
        break

plt.show()
```
